```python
import pandas as pd
import matplotlib.pyplot as plt

data_dir = '../data'
```

```python
df = pd.read_csv(data_dir + '/pd_speech_features.csv')
display(df)
print(df.shape)
```

```python
print(df.columns)
```

```python
demographics = df['class']
print(demographics, '\n')

hc_num = len([pp for pp in demographics if pp == 0])
pd_num = len([pp for pp in demographics if pp == 1])
print("number of HC:", hc_num, hc_num/3)
print("number of PD:", pd_num, pd_num/3)
print("TOTALL:", hc_num + pd_num, (hc_num + pd_num)/3)
```

```python
# df[df.columns[:30]].hist(figsize=(8,8))
df[df.columns[2:4]].hist(figsize=(8,4))
plt.tight_layout()
plt.show()
plt.close()
```

```python
for feature in df.columns[:30]:
    plt.figure()
    plt.scatter(df['id'], df[feature])
    plt.title(feature)
    plt.show()
    plt.close()
```

```python
dfHC = df[df['class'] == 0]
dfPD = df[df['class'] == 1]

dfHC_mean = dfHC.mean()
dfPD_mean = dfPD.mean()

print(dfHC_mean.values[:10])
print(dfPD_mean.values[:10])
```

```python
for feature in df.columns[:30]:
    plt.figure()
    plt.scatter(dfHC['id'], dfHC[feature], label='HC')
    plt.axhline(y=dfHC[feature].mean(), color='blue')
    plt.scatter(dfPD['id'], dfPD[feature], label='PD')
    plt.axhline(y=dfPD[feature].mean(), color='orange')
    plt.title(feature)
    plt.xlabel("patient id")
    plt.legend()
    plt.show()
    plt.close()
```

From the gender plot, we see that the HC is skewed towards 0, since it is less than 0.5. We know from the dataset write up that the HC is 41 women and 23 men. We see that 41/(41+23) = 0.64, which skews towards 1, and 23/(41+23) = 0.36, which skews towards 0. Therefore, we conclude that 0 is men and 1 is women.
